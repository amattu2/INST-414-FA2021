```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


from sklearn.model_selection import train_test_split
import sklearn.metrics as metrics
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
x=iris.data
y=iris.target
```

```python
from sklearn  import tree

xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25)

model=tree.DecisionTreeClassifier() #default criterion is gini impurity
#we can use cross validation to choose between entroy and gini (they are hyper parameters)
model.fit(xtrain,ytrain)

model.score(xtest,ytest)
```

```python
from sklearn  import tree

xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25)

model=tree.DecisionTreeClassifier(max_depth=2) 

model.fit(xtrain,ytrain)

print(model.score(xtest,ytest))
tree.plot_tree(model)
```

```python
from sklearn.tree import export_text
r = export_text(model, feature_names=iris['feature_names'])
r
```

```python

from mlxtend.plotting import plot_decision_regions
from sklearn.datasets import make_circles

x,y=make_circles(n_samples=500,noise=.15,factor=.3)
#change noise to see the change in boundries


model=tree.DecisionTreeClassifier()

model.fit(x,y)
ypred=model.predict(x)


plot_decision_regions(x,y,clf=model,legend=2);

#notice all the angles are 90 degrees 
#they are trademarks of decsion tree classfier
#they correspond to questions and splits
model.score(x,y)
```

```python
from mlxtend.plotting import plot_decision_regions
from sklearn.datasets import make_moons

x,y=make_moons(n_samples=100,shuffle=True, noise=.1)


model=tree.DecisionTreeClassifier()
model.fit(x,y)
ypred=model.predict(x)


plot_decision_regions(x,y,clf=model,legend=2);


model.score(x,y)
```
